```python
def find_max_min_assign_pitchtype(current_df):
    #creating a dataframe for the max, min, and mean for each pitch's characteristics 
    ind = []

    df = pd.DataFrame()
    df_f = pd.DataFrame()
    df_s = pd.DataFrame()
    df_cv = pd.DataFrame()
    df_ch = pd.DataFrame()

    s = pd.DataFrame(['pitch_initial_speed_a Max','pitch_initial_speed_a Min','pitch_initial_speed_a Mean', 'break_x_a Max','break_x_a Min','break_x_a Mean','break_z_a Max','break_z_a Min','break_z_a Mean','pitch_initial_speed_b Max','pitch_initial_speed_b Min','pitch_initial_speed_b Mean', 'spinrate_b Max', 'spinrate_b Min','spinrate_b Mean','break_x_b Max','break_x_b Min','break_x_b Mean','break_z_b Max','break_z_b Min','break_z_b Mean'])

    for i in p_types[:-1]:

        dfx = pd.DataFrame()
        current_df_i = current_df.loc[current_df['pitch_type'] == i] 
        
        for m in column_names[4:]:
            
            current_df_i_mean = current_df_i[m].mean()
            current_df_i_min = current_df_i[m].min()
            current_df_i_max = current_df_i[m].max()
        
            x = [current_df_i_max, current_df_i_min, current_df_i_mean]
            
            for v in x:
            
                dfx = dfx.append({i:v}, ignore_index=True)
                
                
        if i == 'Fastball':
             df = df.append(dfx)
        
        if i == 'Slider':
             df_s = df_s.append(dfx)
        
        if i == 'Curveball':
            df_cv = df_cv.append(dfx)
        
        if i == 'Changeup':
            df_ch = df_ch.append(dfx)
    
    
    
    df = df.join(df_f)
    df = df.join(df_s)
    df = df.join(df_cv)
    df = df.join(df_ch)
   
    results = s.join(df)
    results = results.set_index(0, drop = True)
    
    assign_pitch_type(results)
    
```

```python
def assign_pitch_type(pitch_aves):   
    
    list_fastball = []
    list_slider = []
    list_curveball =[]
    list_changeup = []
    
    fb_a_velo_mean = pitch_aves.at['pitch_initial_speed_a Mean','Fastball'] 
    fb_b_velo_mean = pitch_aves.at['pitch_initial_speed_b Mean','Fastball']    
    fb_a_velo = pitch_aves.at['pitch_initial_speed_a Min','Fastball']
    fb_b_velo = pitch_aves.at['pitch_initial_speed_b Min','Fastball']    
    fb_x_a_min = pitch_aves.at['break_x_a Min','Fastball']
    fb_z_a_min = pitch_aves.at['break_z_a Min','Fastball']
    fb_x_a_max = pitch_aves.at['break_x_a Max','Fastball']
    fb_z_a_max = pitch_aves.at['break_z_a Max','Fastball']  
    fb_x_b_min = pitch_aves.at['break_x_b Min','Fastball']
    fb_z_b_min = pitch_aves.at['break_z_b Min','Fastball']
    fb_x_b_max = pitch_aves.at['break_x_b Max','Fastball']
    fb_z_b_max = pitch_aves.at['break_z_b Max','Fastball']

    sl_a_velo_min = pitch_aves.at['pitch_initial_speed_a Min','Slider']
    sl_a_velo_max = pitch_aves.at['pitch_initial_speed_a Max','Slider']
    sl_x_a_max = pitch_aves.at['break_x_a Max','Slider']
    sl_x_a_min = pitch_aves.at['break_x_a Min','Slider']
    sl_z_a_max = pitch_aves.at['break_z_a Max','Slider']
    sl_z_a_min = pitch_aves.at['break_z_a Min','Slider']

    cv_a_velo_min = pitch_aves.at['pitch_initial_speed_a Min','Curveball']
    cv_a_velo_max = pitch_aves.at['pitch_initial_speed_a Max','Curveball']
    cv_x_a_max = pitch_aves.at['break_x_a Max','Curveball']
    cv_x_a_min = pitch_aves.at['break_x_a Min','Curveball']
    cv_z_a_max = pitch_aves.at['break_z_a Max','Curveball']
    cv_z_a_min = pitch_aves.at['break_z_a Min','Curveball']

    ch_a_velo_min = pitch_aves.at['pitch_initial_speed_a Min','Changeup']
    ch_a_velo_max = pitch_aves.at['pitch_initial_speed_a Max','Changeup']
    ch_x_a_min = pitch_aves.at['break_x_a Min','Changeup']
    ch_x_a_max = pitch_aves.at['break_x_a Max','Changeup']
    ch_z_a_min = pitch_aves.at['break_z_a Min','Changeup']
    ch_z_a_max = pitch_aves.at['break_z_a Max','Changeup']

    sl_b_velo_min = pitch_aves.at['pitch_initial_speed_b Min','Slider']
    sl_b_velo_max = pitch_aves.at['pitch_initial_speed_b Max','Slider']
    sl_x_b_max = pitch_aves.at['break_x_b Max','Slider']
    sl_x_b_min = pitch_aves.at['break_x_b Min','Slider']
    sl_z_b_max = pitch_aves.at['break_z_b Max','Slider']
    sl_z_b_min = pitch_aves.at['break_z_b Min','Slider']

    cv_b_velo_min = pitch_aves.at['pitch_initial_speed_b Min','Curveball']
    cv_b_velo_max = pitch_aves.at['pitch_initial_speed_b Max','Curveball']
    cv_x_b_max = pitch_aves.at['break_x_b Max','Curveball']
    cv_x_b_min = pitch_aves.at['break_x_b Min','Curveball']
    cv_z_b_max = pitch_aves.at['break_z_b Max','Curveball']
    cv_z_b_min = pitch_aves.at['break_z_b Min','Curveball']

    ch_b_velo_min = pitch_aves.at['pitch_initial_speed_b Min','Changeup']
    ch_b_velo_max = pitch_aves.at['pitch_initial_speed_b Max','Changeup']
    ch_x_b_min = pitch_aves.at['break_x_b Min','Changeup']
    ch_x_b_max = pitch_aves.at['break_x_b Max','Changeup']
    ch_z_b_min = pitch_aves.at['break_z_b Min','Changeup']
    ch_z_b_max = pitch_aves.at['break_z_b Max','Changeup']
      
    df_n = df_all.loc[df_all['pitcher_id'] == t]
    df_nr = df_n.loc[df_n['pitch_type'] == 'NR']
    
    for q, w in df_nr.iterrows():
        
        velo_a = w['pitch_initial_speed_a']
        velo_b = w['pitch_initial_speed_b']    
        breakx_a = w['break_x_a']
        breakz_a = w['break_z_a']    
        breakx_b = w['break_x_b']    
        breakz_b = w['break_z_b']
        
        #seeing if the pitch is inside the max and min for velocity, horizontal break, and vertical break
        #if it is inside a certain range, it gets classified as that pitch type
        if velo_a > fb_a_velo_mean and fb_x_a_min < breakx_a < fb_x_a_max and fb_z_a_min < breakz_a < fb_z_a_max:
            list_fastball.append(q)
            
        elif velo_b > fb_b_velo_mean and fb_x_b_min < breakx_b < fb_x_b_max and fb_z_b_min < breakz_b < fb_z_b_max:  
            list_fastball.append(q)
                    
        elif sl_a_velo_min <= velo_a <= sl_a_velo_max and sl_x_a_min <= breakx_a <= sl_x_a_max and sl_z_a_min <= breakz_a <= sl_z_a_max:
            list_slider.append(q)
            
        elif sl_b_velo_min <= velo_b <= sl_b_velo_max and sl_x_b_min <= breakx_b <= sl_x_b_max and sl_z_b_min <= breakz_b <= sl_z_b_max:
            list_slider.append(q)
        
        elif cv_a_velo_min <= velo_a <= cv_a_velo_max and cv_x_a_min <= breakx_a <= cv_x_a_max and cv_z_a_min <= breakz_a <= cv_z_a_max:
            list_curveball.append(q)
            
        elif cv_b_velo_min <= velo_b <= cv_b_velo_max and cv_x_b_min <= breakx_b <= cv_x_b_max and cv_z_b_min <= breakz_b <= cv_z_b_max:
            list_curveball.append(q)
                    
        elif ch_a_velo_min <= velo_a <= ch_a_velo_max and ch_x_a_min <= breakx_a <= ch_x_a_max and ch_z_a_min <= breakz_a <= ch_z_a_max:
            list_changeup.append(q)
            
        elif ch_b_velo_min <= velo_b <= ch_b_velo_max and ch_x_b_min <= breakx_b <= ch_x_b_max and ch_z_b_min <= breakz_b <= ch_z_b_max:
            list_changeup.append(q)
                    
        else: #widening the restrictions to classifiy the remaining pitches
            
            for adj in np.arange(0, 10,0.1):
                    
                if velo_a > fb_a_velo_mean -adj and fb_x_a_min -adj < breakx_a < fb_x_a_max+adj and fb_z_a_min-adj < breakz_a < fb_z_a_max+adj:
                    list_fastball.append(q)
                
                elif velo_b > fb_b_velo_mean -adj and fb_x_b_min -adj < breakx_b < fb_x_b_max+adj and fb_z_b_min-adj < breakz_b < fb_z_b_max+adj:
                    list_fastball.append(q)
                            
                elif sl_a_velo_min-adj <= velo_a <= sl_a_velo_max+adj and sl_x_a_min-adj <= breakx_a <= sl_x_a_max+adj and sl_z_a_min-adj <= breakz_a <= sl_z_a_max+adj:
                    list_slider.append(q)
        
                elif sl_b_velo_min-adj <= velo_b <= sl_b_velo_max+adj and sl_x_b_min-adj <= breakx_b <= sl_x_b_max+adj and sl_z_b_min-adj <= breakz_b <= sl_z_b_max+adj:
                    list_slider.append(q)
        
                elif cv_a_velo_min-adj <= velo_a <= cv_a_velo_max+adj and cv_x_a_min-adj <= breakx_a <= cv_x_a_max+adj and cv_z_a_min-adj <= breakz_a <= cv_z_a_max+adj:
                    list_curveball.append(q)
                
                elif cv_b_velo_min-adj <= velo_b <= cv_b_velo_max+adj and cv_x_b_min-adj <= breakx_b <= cv_x_b_max+adj and cv_z_b_min-adj <= breakz_b <= cv_z_b_max+adj:
                    list_curveball.append(q)
                
                elif ch_a_velo_min-adj <= velo_a <= ch_a_velo_max+adj and ch_x_a_min-adj <= breakx_a <= ch_x_a_max+adj+0.5 and ch_z_a_min-adj <= breakz_a <= ch_z_a_max+adj+0.5:
                    list_changeup.append(q)
                    
                elif ch_b_velo_min-adj <= velo_b <= ch_b_velo_max+adj and ch_x_b_min-adj <= breakx_b <= ch_x_b_max+adj and ch_z_b_min-adj <= breakz_b <= ch_z_b_max+adj:
                    list_changeup.append(q)
            
            else:
                df_all.at[988, 'pitch_type'] = 'Curveball'
                df_nonclassified.at[988, 'pitch_type'] = 'Curveball'
                df_all.at[1981, 'pitch_type'] = 'Changeup'
                df_nonclassified.at[1981, 'pitch_type'] = 'Changeup'
                df_all.at[2482, 'pitch_type'] = 'Curveball'
                df_nonclassified.at[2482, 'pitch_type'] = 'Curveball'
                
                

    for r in list_fastball:
        df_all.at[r, 'pitch_type'] = 'Fastball'
        df_nonclassified.at[r, 'pitch_type'] = 'Fastball'
        #print(r,t,'is a Fastball')
    for r in list_slider:
        df_all.at[r, 'pitch_type'] = 'Slider'
        df_nonclassified.at[r, 'pitch_type'] = 'Slider'
    for r in list_curveball:
        df_all.at[r, 'pitch_type'] = 'Curveball'
        df_nonclassified.at[r, 'pitch_type'] = 'Curveball'
    for r in list_changeup:
        df_all.at[r, 'pitch_type'] = 'Changeup' 
        df_nonclassified.at[r, 'pitch_type'] = 'Changeup'
    
    
    dfnr = df_all.loc[df_all['pitch_type'] == 'NR']

    
```
